## 02. Create Bar Chart

## 03. Create Histogram and Scatterplot

```python
# Creating a histogram
df_combined ['prices'].plot.hist(bins = 25)
```

```python
# Graph above doesn't make sense, create scatterplot to identify outliers
sns.scatterplot(x = 'prices', y = 'prices',data = df_combined)
# Can see we have a few outliers that are skewing the data
```

```python
# identifying prices over 100
df_combined.loc[df_combined['prices'] > 100]
```

```python
# turning outliers into NaNs 
df_combined.loc[df_combined['prices'] >100, 'prices'] = np.nan
```

```python
# Checking max prices now that prices over 100 have been excluded
df_combined['prices'].max()
```

```python
# Re-creating histogram

hist = df_combined['prices'].plot.hist(bins=70)
```

```python
# Export histogram chart
hist.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'hist_prices_combined.png'))
```

```python
#Re-create scatterplot chart
scatterplot = sns.scatterplot(x = 'prices', y = 'prices',data = df_combined)
```

```python
# Export scatterplot chart
scatterplot.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'scatterplot_prices_combined.png'))
```

## 04. Create Line Chart

```python
#Sampling data 70/30 split
# Create a list holding True/False values to the test np.random.rand()<== 0.7
np.random.seed(4)
dev = np.random.rand(len(df_combined)) <= 0.7
```

```python
np.random.rand(10)
```

```python
# Store 70% of the sample in the dataframe big

big = df_combined[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = df_combined[~dev]
```

```python
# Checking for accuracy 
len(df_combined)
```

```python
len(big) + len(small)
```

```python
# reducing samples to only those column necessary for the chart, prices and orders_day_of_week

df_2 = small[['orders_day_of_week','prices']]
```

```python
# creating line chart from sampling

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export scatterplot chart
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_ords_prices_combined.png'))
```
