## 03. Find aggregated mean of the order_number column group by department_id

```python
# Group by product_name
df_ords_prods_merged.groupby('product_name')
```

```python
# Split data into groups by department_id and then calculate the mean values for the order_number column
df_ords_prods_merged.groupby('department_id').agg({'order_number': ['mean']})
```

### 04. The subset has a lot more variability in the mean values. For instance, the lowest mean was 11.29 department_id 4 and the highest was 19.46 department_id 16. With the full dataset the lowest mean is 15.21 department_id 5 and the highest is 22.9 department_id 21.

## 05.  Create a loyalty flag for existing customers

```python
# Group data by user_id column, apply transform function on the order_number column to generate the maximum orders for each user and create a new column max_order to place results of aggregation
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merged.head(30)
```

```python
# Create flag to assign loyalty labels based on max order value
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequencies of loyalty flags for accuracy
df_ords_prods_merged['loyalty_flag'].value_counts(dropna= False)
```

## 06. Determime if there's a difference between the spending habits of the three types of customers identified.

```python
# Group by prices
df_ords_prods_merged.groupby('prices')
```

```python
#Split the data into groups by loyalty_flag and calculate mean each prices column
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean']})
```

## 07. Create a spending flag for each user based on averaging price across all orders

```python
# Group data by user_id column, apply transform function on the prices column to generate mean_price aggregation results
df_ords_prods_merged['mean_price'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merged.head(30)
```

```python
# Create flag to assign spending labels to user_id based on mean product price
df_ords_prods_merged.loc[df_ords_prods_merged['mean_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['mean_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# Checking frequencies of spending flags for accurary 
df_ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merged.head(30)
```

## 08. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_last_order” column.

```python
# Group data by user_id column, apply transform function on the days_since_last_order column to generate median_days_last_order aggregation results
df_ords_prods_merged['median_days_last_order'] = df_ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
df_ords_prods_merged.head(30)
```

```python
# Create flag to assign frequency labels to user_id based on median days since last order
df_ords_prods_merged.loc[df_ords_prods_merged['median_days_last_order'] > 20, 'frequency_flag'] = 'Non-Frequent Customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_days_last_order'] > 10 & (df_ords_prods_merged['median_days_last_order'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_days_last_order'] <= 10 , 'frequency_flag'] = 'Frequent Customer'
```

```python
df_ords_prods_merged.head(30)
```

```python
# Check frequencies of frequency flags for accuracy
df_ords_prods_merged['frequency_flag'].value_counts(dropna=False)
```

```python
# Examining the null values
print(df_ords_prods_merged[df_ords_prods_merged.frequency_flag.isnull()])
```

#### The 5 null values are all new customers, so they only have one order hence why the median days since last prior order cannot be calculated.
