## 03. Data Quality and Consistency

```python
# Check data types for df_customers
df_customers.dtypes
```

```python
# Check for mixed data types

for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

#### Identified that first_name is a mixed data type, will update first_name to string

```python
# Updating first_name to a string
df_customers['first_name'] = df_customers['first_name'].astype('str')
```

```python
# Rerun checking for mixed data types

for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```
