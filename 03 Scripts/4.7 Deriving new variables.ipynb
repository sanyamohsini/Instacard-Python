## 03. Using If Statements

```python
# Define price label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply price label on dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# check max price of data subset 
df['prices'].max()
```

## 04. If-Statements with the loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Check value counts
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Repeat process on entire dataframe
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

## 05. If-Statements with For-Loops

```python
#Create a new column in ords_prods_merged, find frequency 
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Set loop
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Add result to dataframe
df_ords_prods_merged['busiest_day']= result
```

```python
# Check value counts
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Step 2

```python
# Create new if-statement for loop with new busy day requirements
result_2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
# Create the busiest_days column
df_ords_prods_merged['busiest_days'] = result_2
```

```python
# Check
df_ords_prods_merged.head()
```

## Step 3

```python
# Check value counts of busiest days column
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

#### Checked with original frequency counts. The values add up to the same. The busiest days are 0 - Saturday and 1 - Sunday. The slowest days are 4 - Wednesday and 3- Tuesday.

## Step 4

```python
# Create the busiest_period_of_day column
# Frequency of the values in the 'order_hour_of_day' column
df_ords_prods_merged['order_hour_of_day'].value_counts()
```

#### The 24 hours of the day can be split into three parts. First 'Most orders' hour 10, 11, 14, 15, 13, 12, 16, 9. Next 'Average orders' hour 17, 8, 18, 19, 20, 7, 21,23. Lastly, 'Fewest orders' 23, 6, 0, 1, 5, 2, 4, 3.

```python
# Create a for-loop for busiest hours of day categories
hours = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
# Create the hours column 
df_ords_prods_merged['busiest_period_of_day'] = hours
```

```python
# Check output of 'order_hour_of_day' and 'busiest_period_of_day'
df_ords_prods_merged[['order_hour_of_day','busiest_period_of_day']].head(10)
```
